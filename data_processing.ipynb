```python
import requests
import pandas as pd
import re

# Function to get PubChem CIDs for a given drug name
def get_pubchem_cid(drug_name):
    base_url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/"
    url = f"{base_url}{drug_name}/cids/TXT"
    
    try:
        response = requests.get(url)
        if response.status_code == 200:
            # Clean and return the CID (strip unwanted characters)
            cids = response.text.strip()
            return re.sub(r'\D', '', cids)  # Remove non-digit characters
        else:
            return None
    except Exception as e:
        print(f"Error retrieving PubChem CID for {drug_name}: {e}")
        return None

# List of unique drug names from your earlier steps
unique_drugs = list(set(df['Drug_A'].tolist() + df['Drug_B'].tolist()))

# Prepare a list to hold all CIDs for the drugs
drug_cids = []

# For each drug, get the PubChem CID
for drug in unique_drugs:
    print(f"Processing drug: {drug}")
    cid = get_pubchem_cid(drug)
    
    if cid:
        print(f"Found CID: {cid} for drug: {drug}")
        drug_cids.append(cid)
    else:
        print(f"No CID found for drug: {drug}")

# Create a DataFrame from the list of CIDs
cid_df = pd.DataFrame(drug_cids, columns=['CID'])

# Save the DataFrame to a CSV file
cid_df.to_csv('drug_cids_cleaned.csv', index=False)

# Optionally, print the first few rows to confirm
print(cid_df.head())
```
